## Splitting the dataset.

```python
from sklearn.model_selection import train_test_split
```

```python
train,test = train_test_split(raw_df,test_size=0.2,random_state=0)
```

```python
len(train)
```

Our training dataset has 614 instances.

```python
len(test)
```

Our testing dataset has 154 instances.

```python
import numpy as np
```

```python
train_df = train.copy()
```

Let's create a copy of training dataset that it would not alter the original dataset.

```python
test_df = test.copy()
```

Let's create a copy of testing dataset that it would not alter the original dataset.

```python
train_df.replace({
    'Glucose':{0:np.NaN},
    'BloodPressure':{0:np.NaN},
    'SkinThickness':{0:np.NaN},
    'Insulin':{0:np.NaN},
    'BMI':{0:np.NaN},
},inplace=True)
```

Let's replace the zeroes in some of the columns with NaN as these values coludn't be zero.

```python
test_df.replace({
    'Glucose':{0:np.NaN},
    'BloodPressure':{0:np.NaN},
    'SkinThickness':{0:np.NaN},
    'Insulin':{0:np.NaN},
    'BMI':{0:np.NaN},
},inplace=True)
```

```python
input_cols = raw_df.columns[:-1]
```

```python
input_cols
```

Above are the input columns to be given to our model.

```python
target_col = raw_df.columns[-1]
```

```python
target_col
```

Above are the targets to be predicted.

```python
train_inputs = train_df[input_cols].copy()
```

```python
train_targets = train_df[target_col].copy()
```

```python
test_inputs = test_df[input_cols].copy()
```

```python
test_targets = test_df[target_col].copy()
```

```python
train_df.info()
```

```python
test_df.info()
```

We could see that there are some missing values in some of the columns in both training and testing datasets.

## Imputation

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='median')
```

```python
imputer.fit(train_inputs)
```

```python
train_inputs = imputer.transform(train_inputs)
```

```python
test_inputs = imputer.transform(test_inputs)
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(train_inputs)
```

```python
train_inputs = scaler.transform(train_inputs)
```

```python
test_inputs = scaler.transform(test_inputs)
```

## RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(random_state=0,n_jobs=-1)
```

```python
rf_model.fit(train_inputs,train_targets)
```

```python
rf_model.score(train_inputs,train_targets)
```

Above is the training accuracy of our rf_model.

```python
rf_model.score(test_inputs,test_targets)
```

Above is the testing accuracy of our model.
